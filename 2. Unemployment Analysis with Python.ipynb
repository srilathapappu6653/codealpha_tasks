```python
# PART 1: IMPORT, LOAD DATA & DATA CLEANING
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot Style
plt.style.use("ggplot")
sns.set_style("whitegrid")
```

```python
#  Load Datasets
df1 = pd.read_csv(r"D:\codealpha_tasks\Unemployment in India.csv")
df2 = pd.read_csv(r"D:\codealpha_tasks\Unemployment_Rate_upto_11_2020.csv")

```

```python
#  Display First Five Rows
print("First Dataset")
print(df1.head())

print("\nSecond Dataset")
print(df2.head())
```

```python
#  Dataset Shape
print("\nFirst Dataset Shape:", df1.shape)
print("Second Dataset Shape:", df2.shape)

```

```python
#  Column Names
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()

print("\nFirst Dataset Columns")
print(df1.columns)

print("\nSecond Dataset Columns")
print(df2.columns)

```

```python
#  Dataset Information
print("\nFirst Dataset Info")
print(df1.info())

print("\nSecond Dataset Info")
print(df2.info())
```

```python
# Missing Values
print("\nMissing Values in First Dataset")
print("\nMissing Values in Dataset 2")
print(df1.isnull().sum())
print(df2.isnull().sum())
```

```python
#  Remove Duplicate Rows
df1.drop_duplicates(inplace=True)
df2.drop_duplicates(inplace=True)

print("\nDuplicates Removed Successfully")
```

```python
#  Rename Columns
df1.columns = [
    "Region",
    "Date",
    "Frequency",
    "Estimated Unemployment Rate",
    "Estimated Employed",
    "Estimated Labour Participation Rate",
    "Area"
]
```

```python
df1.head()
```

```python
df2.columns = [
    "Region",
    "Date",
    "Frequency",
    "Estimated Unemployment Rate",
    "Estimated Employed",
    "Estimated Labour Participation Rate",
    "State",
    "Longitude",
    "Latitude"
]
```

```python
df2.head()
```

```python
#  Convert Date Column
df1["Date"] = pd.to_datetime(df1["Date"], errors="coerce")
df2["Date"] = pd.to_datetime(df2["Date"], errors="coerce")
```

```python
print(df1.dtypes)
print(df2.dtypes)
```

```python
#  Summary Statistics
print("\nSummary Statistics - Dataset 1")
print(df1.describe())

print("\nSummary Statistics - Dataset 2")
print(df2.describe())

```

```python
#  Unique Values
print("\nNumber of Regions :", df1["Region"].nunique())
print("Number of Areas :", df1["Area"].nunique())
```

```python
#  Visualization 1
# Distribution of Unemployment Rate
plt.figure(figsize=(8,5))
sns.histplot(
    df1["Estimated Unemployment Rate"],
    bins=20,
    kde=True,
    color="royalblue"
)

plt.title("Distribution of Estimated Unemployment Rate")
plt.xlabel("Estimated Unemployment Rate (%)")
plt.ylabel("Count")
plt.show()
```

```python
#  Visualization 2
# Box Plot
plt.figure(figsize=(8,5))
sns.boxplot(
    y=df1["Estimated Unemployment Rate"],
    color="orange"
)

plt.title("Box Plot of Estimated Unemployment Rate")
plt.show()
```

```python
#  Visualization 3
# Average Unemployment Rate by Region
plt.figure(figsize=(10,6))

region_avg = df1.groupby("Region")[
    "Estimated Unemployment Rate"
].mean().sort_values()

sns.barplot(
    x=region_avg.index,
    y=region_avg.values,
    hue=region_avg.index,
    palette="viridis",
    legend=False
)

plt.xticks(rotation=45)
plt.xlabel("Region")
plt.ylabel("Average Unemployment Rate")
plt.title("Average Unemployment Rate by Region")
plt.show()
```

```python
print("\nPART 1 COMPLETED SUCCESSFULLY")
```

```python
#part2
# Top 10 Areas with Highest Average Unemployment Rate

top10 = (
    df1.groupby("Area")["Estimated Unemployment Rate"]
       .mean()
       .sort_values(ascending=False)
       .head(10)
)

plt.figure(figsize=(10,6))

sns.barplot(
    x=top10.values,
    y=top10.index,
    hue=top10.index,
    palette="Reds_r",
    legend=False
)

plt.title("Top 10 Areas with Highest Average Unemployment Rate")
plt.xlabel("Average Unemployment Rate")
plt.ylabel("Area")
plt.show()
```

```python
# Top 10 Areas with Lowest Average Unemployment Rate

bottom10 = (
    df1.groupby("Area")["Estimated Unemployment Rate"]
       .mean()
       .sort_values()
       .head(10)
)

plt.figure(figsize=(10,6))

sns.barplot(
    x=bottom10.values,
    y=bottom10.index,
    hue=bottom10.index,
    palette="Greens",
    legend=False
)

plt.title("Top 10 Areas with Lowest Average Unemployment Rate")
plt.xlabel("Average Unemployment Rate")
plt.ylabel("Area")

plt.show()
```

```python
# Pie Chart of Region Distribution
region_count = df1["Region"].value_counts()

plt.figure(figsize=(8,8))
plt.pie(
    region_count,
    labels=region_count.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Distribution of Regions")
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(8,6))
sns.scatterplot(
    x="Estimated Employed",
    y="Estimated Unemployment Rate",
    hue="Region",
    data=df1
)

plt.title("Employment vs Unemployment Rate")
plt.show()
```

```python
# Correlation Heatmap
numeric = df1.select_dtypes(include=["float64","int64"])

plt.figure(figsize=(8,6))
sns.heatmap(
    numeric.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()


```

```python
print("Part 2 Completed Successfully")
```

```python
# PART 3: COVID-19 ANALYSIS & CONCLUSION
# COVID-19 Impact Analysis
covid_data = df1[df1["Date"].dt.year == 2020]

plt.figure(figsize=(14,6))

sns.lineplot(
    data=covid_data,
    x="Date",
    y="Estimated Unemployment Rate",
    hue="Region"
)

plt.title("COVID-19 Impact on Unemployment Rate (2020)")
plt.xlabel("Date")
plt.ylabel("Estimated Unemployment Rate (%)")
plt.xticks(rotation=45)

plt.show()
```

```python
# Monthly Trend
df1["Month"] = df1["Date"].dt.month_name()

month_order = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]

monthly = df1.groupby("Month")["Estimated Unemployment Rate"].mean()

monthly = monthly.reindex(month_order)

plt.figure(figsize=(12,6))

sns.lineplot(
    x=monthly.index,
    y=monthly.values,
    marker="o"
)

plt.xticks(rotation=45)

plt.title("Average Monthly Unemployment Rate")
plt.xlabel("Month")
plt.ylabel("Average Unemployment Rate")

plt.show()
```

```python
# Highest Unemployment Region
highest_region = (
    df1.groupby("Region")["Estimated Unemployment Rate"]
       .mean()
       .sort_values(ascending=False)
)

print("\nAverage Unemployment Rate by Region")
print(highest_region)

```

```python
# Highest Unemployment Region
highest_area = (
    df1.groupby("Area")["Estimated Unemployment Rate"]
       .mean()
       .sort_values(ascending=False)
       .head(10)
)

print("\nTop 10 Areas with Highest Average Unemployment")
print(highest_area)

```
